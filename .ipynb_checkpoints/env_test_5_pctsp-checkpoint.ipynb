## 4. How to compute cost

```python
# load pretrained result
def load_sample_data(loadpath):
    with open(loadpath, 'rb') as f:
        return pickle.load(f)
    
loadpath = "sample_data/pctsp100_det.pkl"
pctsp100_samples = load_sample_data(loadpath)
print(len(pctsp100_samples))

# batch_size = 1
single_sample = pctsp100_samples[-1]
print(single_sample['cost'])
single_sample_inputs = {'loc': single_sample['loc'],
                        'deterministic_prize': single_sample['deterministic_prize'],
                        'stochastic_prize': single_sample['stochastic_prize'],
                        'depot': single_sample['depot'],
                        'penalty': single_sample['penalty']}
```

```python
cost = pctsp_env.compute_cost(single_sample_inputs, single_sample['solution'])
print(cost, cost.shape)
```

```python
single_sample['solution']
```

## 5. Visualize

```python
for i in range(len(pctsp100_samples)):
    curr_sample = pctsp100_samples[i]
    curr_sample_inputs = {'loc': curr_sample['loc'],
                          'deterministic_prize': curr_sample['deterministic_prize'],
                          'stochastic_prize': curr_sample['stochastic_prize'],
                          'depot': curr_sample['depot'],
                          'penalty': curr_sample['penalty']}
    curr_sample_selected_idxs = curr_sample['solution']
    curr_sample_cost = curr_sample['cost']
    print("Cost: {:.2f}".format(curr_sample_cost[0]))
    pctsp_env.plot(curr_sample_inputs, 
                    curr_sample_selected_idxs, 
                    show_prize_and_penalty=True,
                    savepath=None)
    print("==================")
```
