## 4. How to compute cost

```python
# load pretrained result
def load_sample_data(loadpath):
    with open(loadpath, 'rb') as f:
        return pickle.load(f)
    
loadpath = "sample_data/op100_dist.pkl"
op100_samples = load_sample_data(loadpath)
print(len(op100_samples))

# batch_size = 1
single_sample = op100_samples[-1]
print(single_sample['cost'])
single_sample_inputs = {'loc': single_sample['loc'],
                        'prize': single_sample['prize'],
                        'depot': single_sample['depot'],
                        'max_length': single_sample['max_length']}
```

```python
# The goal is to maximize the prize
# cost = -prize
cost = op_env.compute_cost(single_sample_inputs, single_sample['solution'])
print(cost, cost.shape)
```

```python
single_sample['solution']
```

## 5. Visualize

```python
for i in range(len(op100_samples)):
    curr_sample = op100_samples[i]
    curr_sample_inputs = {'loc': curr_sample['loc'],
                        'prize': curr_sample['prize'],
                        'depot': curr_sample['depot'],
                        'max_length': curr_sample['max_length']}
    curr_sample_selected_idxs = curr_sample['solution']
    curr_sample_cost = curr_sample['cost']
    print("Cost: {:.2f}".format(curr_sample_cost[0]))
    op_env.plot(curr_sample_inputs, 
                       curr_sample_selected_idxs, 
                       show_prize=True,
                       savepath=None)
    print("==================")
```
